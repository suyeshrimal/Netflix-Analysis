#### Q. In which year highest number of TV shows and movies were released? Show with the bar Graph.

```python
df['Release_Date'] = df['Release_Date'].str.strip()
```

```python
df['Release_Date']=pd.to_datetime(df['Release_Date'],format='%B %d, %Y')
df.drop(columns='Date_N')
```

```python
highest_number_of_TV_shows_and_movies=df['Release_Date'].dt.year.value_counts()
highest_number_of_TV_shows_and_movies
```

```python
highest_number_of_TV_shows_and_movies.plot(kind='bar')
```
